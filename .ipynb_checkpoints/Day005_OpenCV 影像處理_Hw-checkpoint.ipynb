```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('data/lena.png')
```

### 第一大題 (OpenCV 標註訓練)

1. 將 Lena 的臉蛋用紅框框起來
2. 畫出紅匡中的對角線
3. 標注紅匡左上角的座標
4. 利用 subplot 將三張圖片繪製出來

```python
# 利用 subplot 將三張圖片繪製出來

plt.figure(figsize = (15,10))
# 獲得用紅匡標注的圖片
""" Your Code Here: """
img_rectangle = img.copy()
plt.subplot(1, 3, 1)
plt.imshow(cv2.rectangle(img_rectangle, (60, 40), (420, 510), (0, 0, 255), 3)[:,:,::-1])
# 獲得用對角線標注的圖片
""" Your Code Here: """
img_line = img.copy()
plt.subplot(1, 3, 2)
plt.imshow(cv2.line(img_line, (60, 40), (420, 510), (0, 0, 255), 3)[:,:,::-1])
# 獲得具有座標標注的圖片
""" Your Code Here: """
img_text = img.copy()
plt.subplot(1, 3, 3)
plt.imshow(cv2.putText(img_text, '(60, 40)', (60, 40), 0, 1, (0, 0, 255), 2)[:,:,::-1])
```

### 第二大題 (繪製 Histogram)

1. 獲得 3 個 channel (RGB) 的 histogram
2. 利用 subplot 繪製出來

```python
# 繪製 Histogram
""" Your Code Here: """
plt.figure(figsize = (8, 14))
for i in range(3):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.subplot(3, 1, i + 1, title = 'channel {} histogram'.format(i))
    plt.bar(range(0, 256), hist[:, 0])
plt.show()
```
